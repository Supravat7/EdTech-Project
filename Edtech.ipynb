```python
#rename the columns
ed = ed.rename(columns={'Course name':'course_name', 'Mode Of program':'mode_of_program',
       'Course Duration':'course_duration',
       'Certification Provided':'certification_provided', 'Practice Assignments Included':'practice_assignments_included',
       'Doubt Sessions':'doubt_session', 'Internship Provided':'internship_provided', 'Course rating':'course_rating',
       'Placement Assistance':'placement_assistance', 'Recorded videos Provided':'recorded_videos',
       'Trainer Grade':'trainer_grade', 'Price course':'course_price'})
```

```python
from sklearn.impute import SimpleImputer

#mean imputation for missing values in the data of continuous values.
mean_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
ed['course_price'] = pd.DataFrame(mean_imputer.fit_transform(ed[['course_price']]))

#Median imputation for missing values in the data of continuous values.

median_imputer = SimpleImputer(missing_values=np.nan, strategy='median')
ed['course_duration'] = pd.DataFrame(median_imputer.fit_transform(ed[['course_duration']]))
ed['course_rating'] = pd.DataFrame(median_imputer.fit_transform(ed[['course_rating']]))

ed.isnull().sum()
```

```python
# Studentized Residuals = Residual/standard deviation of residuals
# index 86 is showing high influence so we can exclude that entire row
ed_new = ed.drop(ed.index[[84]])
ed_new
```

```python
# Check for Colinearity to decide to remove a variable using VIF
# Assumption: VIF > 10 = colinearity
# calculating VIF's values of independent variables

rsq_i = smf.ols('institute ~ course_name + mode_of_program + course_duration + Country + City + certification_provided + practice_assignments_included + doubt_session + internship_provided + course_rating + placement_assistance + recorded_videos + trainer_grade', data=ed).fit().rsquared
vif_i = 1/(1 - rsq_i)

rsq_cn = smf.ols('course_name ~ institute + mode_of_program + course_duration + Country + City + certification_provided + practice_assignments_included + doubt_session + internship_provided + course_rating + placement_assistance + recorded_videos + trainer_grade', data=ed).fit().rsquared
vif_cn = 1/(1 - rsq_cn)

rsq_mp = smf.ols('mode_of_program ~ institute + course_name + course_duration + Country + City + certification_provided + practice_assignments_included + doubt_session + internship_provided + course_rating + placement_assistance + recorded_videos + trainer_grade', data=ed).fit().rsquared
vif_mp = 1/(1 - rsq_mp)

rsq_cd = smf.ols('course_duration ~ institute + course_name + mode_of_program + Country + City + certification_provided + practice_assignments_included + doubt_session + internship_provided + course_rating + placement_assistance + recorded_videos + trainer_grade', data=ed).fit().rsquared
vif_cd = 1/(1 - rsq_cd)

rsq_co = smf.ols('Country ~ institute + course_name + mode_of_program + course_duration + City + certification_provided + practice_assignments_included + doubt_session + internship_provided + course_rating + placement_assistance + recorded_videos + trainer_grade', data=ed).fit().rsquared
vif_co = 1/(1 - rsq_co)

rsq_ci = smf.ols('City ~ institute + course_name + mode_of_program + course_duration + Country + certification_provided + practice_assignments_included + doubt_session + internship_provided + course_rating + placement_assistance + recorded_videos + trainer_grade', data=ed).fit().rsquared
vif_ci = 1/(1 - rsq_ci)

rsq_cp = smf.ols('certification_provided ~ institute + course_name + mode_of_program + course_duration + Country + City + practice_assignments_included + doubt_session + internship_provided + course_rating + placement_assistance + recorded_videos + trainer_grade', data=ed).fit().rsquared
vif_cp = 1/(1 - rsq_cp)

rsq_pai = smf.ols('practice_assignments_included ~ institute + course_name + mode_of_program + course_duration + Country + City + certification_provided + doubt_session + internship_provided + course_rating + placement_assistance + recorded_videos + trainer_grade', data=ed).fit().rsquared
vif_pai = 1/(1 - rsq_pai)

rsq_ds = smf.ols('doubt_session ~ institute + course_name + mode_of_program + course_duration + Country + City + certification_provided + practice_assignments_included+ internship_provided + course_rating + placement_assistance + recorded_videos + trainer_grade', data=ed).fit().rsquared
vif_ds = 1/(1 - rsq_ds)

rsq_ip = smf.ols('internship_provided ~ institute + course_name + mode_of_program + course_duration + Country + City + certification_provided + practice_assignments_included + doubt_session + course_rating + placement_assistance + recorded_videos + trainer_grade', data=ed).fit().rsquared
vif_ip = 1/(1 - rsq_ip)

rsq_cr = smf.ols('course_rating ~ institute + course_name + mode_of_program + course_duration + Country + City + certification_provided + practice_assignments_included + doubt_session + internship_provided + placement_assistance + recorded_videos + trainer_grade', data=ed).fit().rsquared
vif_cr = 1/(1 - rsq_cr)

rsq_pa = smf.ols('placement_assistance ~ institute + course_name + mode_of_program + course_duration + Country + City + certification_provided + practice_assignments_included + doubt_session + internship_provided + course_rating + recorded_videos + trainer_grade', data=ed).fit().rsquared
vif_pa = 1/(1 - rsq_pa)

rsq_rv = smf.ols('recorded_videos ~ institute + course_name + mode_of_program + course_duration + Country + City + certification_provided + practice_assignments_included + doubt_session + internship_provided + course_rating + placement_assistance + trainer_grade', data=ed).fit().rsquared
vif_rv = 1/(1 - rsq_rv)

rsq_tg = smf.ols('trainer_grade ~ institute + course_name + mode_of_program + course_duration + Country + City + certification_provided + practice_assignments_included + doubt_session + internship_provided + course_rating + placement_assistance + recorded_videos', data=ed).fit().rsquared
vif_tg = 1/(1 - rsq_tg)
```

```python
# Splitting the data into train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

```python
#now lets build a model using fitting Multilinear Regression to the training set
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
#prediction for the test results
y_pred = lr.predict(X_test)
y_pred
```

```python
# Finding the MAE, MSE, RMSE score
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#for comarision to check whether the y_test is good or y_pred is good
#we will use r^2 score from the sklearn package
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred) 
score
```

```python
print('Train Score: ', lr.score(X_train, y_train))  
print('Test Score: ', lr.score(X_test, y_test)) 
```

```python
#now lets build a model using fitting Random Forest Regression to the training set
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(max_depth = 5, random_state = 42, criterion = 'squared_error')

rf_model.fit(X_train, y_train)

preds = rf_model.predict(X_test)

print('Random Forest: ', r2_score(y_test, preds))
```

```python
pred = rf_model.predict(X_train)
r2_score(y_train,pred)
```

```python
#now lets build a model using fitting Decision Tree Regression to the training set
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)

preds = dt.predict(X_test)

r2_score(y_test, preds)
```

```python
pred = dt.predict(X_train)
r2_score(y_train, pred)
```
